### Conversion of Datatypes 
The survey data is from 2015,We assume that respondents who marked 'present' for 'DOL' (Date of Leaving) actually left the company by 2024-02-17. We'll replace the 'present' value in 'DOL' with this date. After that, we'll convert the 'DOJ' (Date of Joining) and 'DOL' columns to datetime format

```python
df['DOL'].replace('present','2015', inplace= True)
df['DOL'] = pd.to_datetime(df['DOL'])
df['DOJ'] = pd.to_datetime(df['DOJ'])
df.head()
```

```python
cat_ = ['Designation','JobCity', 'Gender','10board','12board','CollegeTier','Degree',
              'Specialization','CollegeCityTier','CollegeState']
for x in cat_:
    df[x] = df[x].astype('category')
```

```python
# describing gender column with Female and Male
df['Gender'].replace({'f':'Female','m':'Male'}, inplace = True)
df.head()
```

From the column descriptions

- '10board', '12board', 'GraduationYear', 'JobCity', 'Domain' cannot have 0 or -1 as inputs. Any 0 or -1 values in these columns should be considered as null values and need to be imputed.

- Some columns describe optional subjects for exams, leading to a large number of -1 (null) values. We'll drop columns where the percentage of -1 values is 80% or more. For the remaining columns, we'll impute -1 values as 0.

```python
df = df.drop(columns = ['MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg'])
df.head()
```

```python
df['10board'] = df['10board'].astype(str)
df['12board'] = df['12board'].astype(str)
df['JobCity'] = df['JobCity'].astype(str)
```

```python
df['10board'] = df['10board'].replace({'0':np.nan})
df['12board'] = df['12board'].replace({'0':np.nan})
df['GraduationYear'] = df['GraduationYear'].replace({0:np.nan})
df['JobCity'] = df['JobCity'].replace({'-1':np.nan})
df['Domain'] = df['Domain'].replace({-1:np.nan})
df['ElectronicsAndSemicon'] = df['ElectronicsAndSemicon'].replace({-1:0})
df['ComputerScience'] = df['ComputerScience'].replace({-1:0})
df['ComputerProgramming'] = df['ComputerProgramming'].replace({-1:np.nan})
```

```python
df['10board'] = df['10board'].astype('category')
df['12board'] = df['12board'].astype('category')
df['JobCity'] = df['JobCity'].astype('category')
```

```python
df
```

```python
# filling null values with mode in categorical columns
df['10board'].fillna(df['10board'].mode()[0], inplace = True)
df['12board'].fillna(df['12board'].mode()[0], inplace = True)
df['GraduationYear'].fillna(df['GraduationYear'].mode()[0], inplace = True)
df['JobCity'].fillna(df['JobCity'].mode()[0], inplace = True)
df
```

```python
# filling null values in numerical columns
df['Domain'].fillna(df['Domain'].median(), inplace = True)
df['ComputerProgramming'].fillna(df['ComputerProgramming'].median(), inplace = True)
df.head()
```

```python
#  correct the string data in columns
def correct_string_data(data):
    '''
    Convert the textual categories to lower case
    and remove the extra spaces.

    '''
    df[data] = df[data].str.lower().str.strip()
```

```python
textual_columns = ['Designation','JobCity','10board','12board','Specialization','CollegeState']
```

```python
for col in textual_columns:
    print(f'Number of unique values in {col} without dissimilarity : {df[col].nunique()}')
```

 number of categories are more, so we can take top 10 from them

```python
def collapsing_categories(df, data):
    for Designation in df[data].unique():
        min_count = df[data].value_counts()[:10].min()
        if df[df[data] == Designation][data].value_counts()[0] < min_count:
            df.loc[df[data] == Designation, data] = 'other'
```

```python
for cols in textual_columns:
    collapsing_categories(df, cols)
```

```python
for cols in textual_columns:
    print('')
    print('Top 10 categories in:', cols)
    print('')
    print(df[cols].value_counts())
    print('')
```

```python
# adding tenure column by subtracting Date of leaving from date of joining
date_ = (df['DOL'] - df['DOJ'])
tenure = np.zeros(len(df))
for i, date in enumerate(date_):
    tenure[i] = round(date.days/365,2)
df['Tenure'] = tenure
```

```python
df = df.drop(df[(df['GraduationYear'] > df['DOJ'].dt.year)].index)
```

```python
## creating function to calculate cdf
def cdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x, y
```

```python
### Removing Outliers
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?

```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
x = np.linspace(0, 100, 100)
y = chi2.pdf(x, df = 4)
plt.plot(x, y)
```

```python
obsr = pd.crosstab(df_.Specialization,df_.Gender)
obsr
```

```python
##Computing Chi2 statistics, p_value and dof

chi2_statistic, chi2_p_value, chi2_dof, chi2_expected = chi2_contingency(obsr)

print("Statistic            :", chi2_statistic)
print('')
print("p value              :", chi2_p_value)
print('')
print("Degrees of freedom   :", chi2_dof)
print('')
print("Expected frequencies array:\n", chi2_expected)
```

```python
### Calculating chi2_critical
confidence_level = 0.95
alpha = 1 - confidence_level

chi2_critical = chi2.ppf(1 - alpha, chi2_dof)

chi2_critical
```

```python
###Chi-squared Test Visualization
x_min = -10
x_max = 100

x = np.linspace(x_min, x_max, 100)
y = chi2.pdf(x, chi2_dof)
plt.xlim(x_min, x_max)
plt.plot(x, y)

chi2_critical_right = chi2_critical

x1 = np.linspace(chi2_critical_right, x_max, 100)
y1 = chi2.pdf(x1, chi2_dof)
plt.fill_between(x1, y1, color='blue')
```

```python
if(chi2_statistic > chi2_critical):
    print("There is not enough evidence to reject the Null Hypothesis")
else:
    print("There is sufficent evidence to reject the Null Hypothesis")
```

```python
if(chi2_p_value < alpha):
    print("There is not enough evidence to reject the Null Hypothesis")
else:
    print("There is sufficent evidence to reject the Null Hypothesis")
```

#### Observations


The analysis suggests a relationship between Gender and Specialization. A Chi-Square test indicates that both categorical variables are dependent, implying that certain specializations or working fields may impose restrictions on candidates based on their gender, possibly due to inherent risks.
